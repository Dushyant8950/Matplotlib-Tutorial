# Write a  Python program to plot two or more lines on same plot with suitable legends of each line.

```python
import matplotlib.pyplot as plt
x1 = [10,20,30]
y1 = [20,40,10]
plt.plot(x1, y1, label = "line 1")
x2 = [10,20,30]
y2 = [40,10,30]
plt.plot(x2, y2, label = "line 2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Two or more lines on same plot")
plt.legend()
plt.show()
```

# Write a Python program to plot two or more lines with legends, different widths and colors.

```python
import matplotlib.pyplot as plt
x1 = [10,20,30]
y1 = [20,40,10]
plt.plot(x1,y1,color="blue",linewidth=2,label="line 1 - width 2")
x2 = [10,20,30]
y2 = [40,10,30]
plt.plot(x2,y2,color="green",linewidth=4,label="line 2 - width 4")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Two or more lines on same plot with diff. color and linewidth")
plt.legend()
plt.show()
```

# Write a  Python program to plot two or more lines with different styles.

```python
import matplotlib.pyplot as plt
x1 = [10,20,30]
y1 = [20,40,10]
plt.plot(x1,y1,color="red",linewidth=2,linestyle="dotted",label="line 1 - width 2, style (dotted)")
x2 = [10,20,30]
y2 = [40,10,30]
plt.plot(x2,y2,color="blue",linewidth=3,linestyle="dashed",label="line 2 - width 3, style (dashed)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Two or more lines on same plot with diff. color, linewidth and linestyle")
plt.legend()
plt.show()
```

# Write a Python program to plot two or more lines and set the line markers.

```python
import matplotlib.pyplot as plt
x = [1,4,5,6,7]
y = [2,6,3,6,3]
plt.plot(x,y,color="blue",linewidth=3,linestyle="solid",marker="o",markerfacecolor="green",markeredgecolor="red",markersize=15)
plt.xlim(0,8)
plt.ylim(0,8)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("two or more lines with markers")
plt.show()
```

# How would you customize the ticks on the X-axis and Y-axis in a Matplotlib plot?

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)

# Customizing ticks
ax.set_xticks([0, 2, 4, 6, 8, 10])
ax.set_xticklabels(['zero', 'two', 'four', 'six', 'eight', 'ten'])
ax.set_yticks([-1, 0, 1])
ax.set_yticklabels(['low', 'zero', 'high'])

plt.show()
```

# How can you add a secondary Y-axis to a plot in Matplotlib?

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax1 = plt.subplots()
ax1.plot(x, y1, 'g-')
ax1.set_ylabel('Sine', color='g')

ax2 = ax1.twinx()
ax2.plot(x, y2, 'b-')
ax2.set_ylabel('Cosine', color='b')

plt.title('Plot with Two Y Axes')
plt.show()
```

# How would you save a Matplotlib figure as an image file?

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Sine Wave')
ax.set_xlabel('x')
ax.set_ylabel('y')

# Save figure
fig.savefig('sine_wave.png')
```
