# Set font properties for title and labels (fontdict) :-

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([80, 85, 90, 95, 100, 105, 110, 115])
y=np.array([240, 250, 260, 270, 280, 290, 300, 310])
font1={'c':'b','size':20}
font2={'c':'r','size':20}
plt.title('Sports Watch Data',fontdict=font1)
plt.xlabel('Average Pulse',fontdict=font2)
plt.ylabel('Calorie Burnage',fontdict=font2)
plt.plot(x,y)
plt.show()
```

```python
''' four famailies for font style
'serif', 'sans-serif', 'monospace', 'fantasy' or 'cursive' '''
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([80, 85, 90, 95, 100, 105, 110, 115])
y=np.array([240, 250, 260, 270, 280, 290, 300, 310])
font1={'family':'cursive','c':'b','size':20}
font2={'family':'fantasy','c':'r','size':20}
plt.title('Sports Watch Data',fontdict=font1)
plt.xlabel('Average Pulse',fontdict=font2)
plt.ylabel('Calorie Burnage',fontdict=font2)
plt.plot(x,y)
plt.show()
```

# Position of title (loc) :-


```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([80, 85, 90, 95, 100, 105, 110, 115])
y=np.array([240, 250, 260, 270, 280, 290, 300, 310])
font1={'family':'sans-serif','c':'b','size':20}
font2={'family':'cursive','c':'r','size':20}
plt.title('Sports Watch Data',fontdict=font1,loc='left') # loc = 'left', 'right', 'center'
plt.xlabel('Average Pulse',fontdict=font2)
plt.ylabel('Calorie Burnage',fontdict=font2)
plt.plot(x,y, c='r', marker='o', mec='k', mfc='y', ms=10)
plt.show()
```

# Multiple ploting (Subplot) :-

```python
import matplotlib.pyplot as plt
import numpy as np
# plot 1 :-
x=np.array([0,1,2,3])
y=np.array([3,1,8,10])
plt.subplot(1,2,1)
plt.title('Plot 1',c='r',family='cursive')
plt.plot(x,y, c='b', ls=':')
# plot 2 :-
x=np.array([0,1,2,3])
y=np.array([6,2,7,11])
plt.subplot(1,2,2)
plt.title('Plot 2',c='r',family='cursive')
plt.plot(x,y, c='g', ls='--')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# plot 1 :-
x=np.array([0,1,2,3])
y=np.array([3,1,8,10])
plt.subplot(2,1,1)
plt.title('Plot 1',c='g',family='fantasy')
plt.plot(x,y, c='b',marker='*', ls=':')
# plot 2 :-
x=np.array([0,1,2,3])
y=np.array([6,2,7,11])
plt.subplot(2,1,2)
plt.title('Plot 2',c='g',family='fantasy')
plt.plot(x,y, c='r',marker='*', ls='--')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# plot 1
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(1,3,1)
plt.title('Plot 1', c='b',family='serif',size=10)
plt.plot(x,y, c='r',ls='-', marker='o')
# plot 2
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,3,2)
plt.title('Plot 2', c='y',family='sans-serif',size=10)
plt.plot(x,y, c='g',ls='--', marker='*')
#plot 3
x=np.array([0,1,2,3])
y=np.array([6,2,7,11])
plt.subplot(1,3,3)
plt.title('Plot 3', c='k',family='monospace',size=10)
plt.plot(x,y, c='y',ls=':', marker='.')
plt.show()
```
